```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('cleaned_weather_data.csv')
```

```python
print("histogramme des temperatures")
plt.hist(data['Temperature'])
plt.xlabel("temperatures")
plt.ylabel("occurences")
plt.show()
```

```python
print("box plox des precipitations")
sns.boxplot(x = data['Precipitation'])
```

```python
temperature_precipitation = data[["Temperature","Precipitation"]] 
sns.regplot(x=data["Temperature"], y=data["Precipitation"], fit_reg=True)
```

```python
print("graphiques temperature et prcipitation en fonction du temps")
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].plot(data["Date"], data["Temperature"], label="Temperature", color="blue")
axs[0].set_title("Temperature Over Time")
axs[0].set_xlabel("Date")
axs[0].set_ylabel("Temperature")


# Plot Precipitation
axs[1].plot(data["Date"], data["Precipitation"], label="Precipitation", color="green")
axs[1].set_title("Precipitation Over Time")
axs[1].set_xlabel("Date")
axs[1].set_ylabel("Precipitation")
plt.tight_layout()


plt.show()
```

```python
print("Affichage de la matrice de correlation : ")
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
correlation_matrixe = data.select_dtypes(include=numerics).corr()
plt.figure(figsize=(8, 6))  # Ajuste la taille de la figure
sns.heatmap(correlation_matrixe, annot=True, cmap="coolwarm", fmt=".2f", cbar=True)
```
